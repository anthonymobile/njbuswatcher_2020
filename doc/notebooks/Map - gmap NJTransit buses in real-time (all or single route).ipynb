```python
import sys
import lib.BusAPI as Buses
import gmaps
import config

gmaps.configure(api_key=config.free_maps_api_key)
```

```python
buses = Buses.parse_xml_getBusesForRouteAll(Buses.get_xml_data('nj', 'all_buses'))
```

```python
buses
```

```python
bus_points = []
for bus in buses:
    if bus.rt == '119':
        bus_points.append(bus)
```

```python
bus_points
```

```python
m = gmaps.Map()
m.add_layer(gmaps.symbol_layer([(float(b.lat), float(b.lon)) for b in bus_points], fill_color='green', stroke_color='green', scale=2))
m
```
